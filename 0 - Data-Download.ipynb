# Download the Retail Rocket Dataset from here
https://www.kaggle.com/datasets/retailrocket/ecommerce-dataset

and upload it to this folder and run the cell below to unzip and put the data on the data/ folder </br>
Rename the download file to "retailrocket-kaggle-data.zip" before or after upload

```python
import shutil
import os

zipfile = "retailrocket-kaggle-data.zip"
DATA_DIR = "./data/"

#Uncompress Datasets zip file on temp folder
os.makedirs(DATA_DIR, exist_ok=True)
shutil.unpack_archive(zipfile, extract_dir=DATA_DIR)
```
